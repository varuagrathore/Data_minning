Univariate

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\dell\Downloads\archive (3)\loan_data_set.csv")
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.histplot(data['Loan_Amount_Term'], kde=True)
plt.title('Histogram')
```

A violin plot is a method to visualize the distribution of data across different categories. It is similar to a box plot, but it also shows the probability density of the data at different values.

```python
# Create violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(data=data['CoapplicantIncome'], color='skyblue')
plt.title('Violin Plot of Example Data')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()
```

```python
# Count the occurrences of each gender
gender_counts = data['Gender'].value_counts()
print(gender_counts)
# Create a bar plot
gender_counts.plot(kind='bar', color=['blue', 'pink'])
```

**Bivariate**

```python
plt.scatter(data['ApplicantIncome'], data['LoanAmount'], color='green', alpha=0.5)
plt.title('Scatter plot of Applicant Income vs Loan Amount')
plt.xlabel('Applicant Income')
plt.ylabel('Loan Amount')
```

```python
# Create a countplot with hue based on loan status for education
sns.countplot(x=data['Education'], hue=data['Loan_Status'])

# Add titles and labels
plt.title('Countplot of Education with Loan Status')
plt.xlabel('Education')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
# Group by Gender and calculate mean of ApplicantIncome
gender_applicant_income = data.groupby('Gender')['ApplicantIncome'].mean()

# Plot the bar plot
gender_applicant_income.plot(kind='bar', color=['blue', 'pink'])

# Add titles and labels
plt.title('Average Applicant Income by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Applicant Income')

# Show the plot
plt.show()
```

The errorbar parameter in Seaborn's barplot function is used to control the display of error bars on the plot. By default, Seaborn calculates and displays a confidence interval around the mean when plotting a bar plot. However, you can set errorbar=None to suppress the display of these error bars if you don't want them to be shown in your plot.

```python
# Create bar plot using Seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x=data['Gender'], y=data['ApplicantIncome'], errorbar=None)

# Add titles and labels
plt.title('Applicant Income by Gender')
plt.xlabel('Gender')
plt.ylabel('Applicant Income')

# Show the plot
plt.show()
```

```python
# Create bar plot using Seaborn
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['Self_Employed'], y=data['ApplicantIncome'])

# Add titles and labels
plt.title('Applicant Income by Gender')
plt.xlabel('Gender')
plt.ylabel('Applicant Income')

# Show the plot
plt.show()
```
